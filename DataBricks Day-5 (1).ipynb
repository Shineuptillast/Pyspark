```
def date_conversion(df):
    date_format=['yyyy-MM-dd','yyyy-dd-MM','yyyy/dd/MM']
    return df.withColumn('value', coalesce(*[ to_date('value', dtf) for dtf in date_format]))
```

```
df.transform(date_conversion).show()
```
